```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import TimeDistributed
from keras.layers import RepeatVector
from keras.layers import RNN
from keras.layers import Dropout

from keras.optimizers import Adam
from keras.optimizers import SGD

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 6]
```

```python
groundTruth = pd.read_csv('./aug22/vectornav.csv')

groundTruth.head()

gt = groundTruth[['field.RPY.x',
 'field.RPY.y',
 'field.RPY.z']]

gt.head()


gt['field.RPY.z'] = gt['field.RPY.z'] - gt.iloc[0,2]

gt.plot()
```

```python

gt[['field.RPY.z']] = gt[['field.RPY.z']]

gtNew = gt

yawAngles = gtNew['field.RPY.z'].values
yawNew = yawAngles

for i in range(len(yawAngles))[1:]:
    #print(i)
    if yawAngles[i] > 180:
        yawAngles[i] = yawAngles[i] - 360

gtNew['field.RPY.z'] = yawAngles

gtNew = gtNew * 3.1415/180
gtNew.plot()
```

```python
ekfData = pd.read_csv('./aug22/rpy_pix_df.csv')
ekfData.head()
```

```python
quat = ekfData[['field.RPY.x',
 'field.RPY.y',
 'field.RPY.z']]

sensors = pd.read_csv('./aug22/aug22_pix.csv')
sensorData = sensors[['field.angular_velocity.x', 'field.angular_velocity.y', 'field.angular_velocity.z', 
                      'field.linear_acceleration.x', 'field.linear_acceleration.y', 'field.linear_acceleration.z']]

quat = quat * 3.1415/180
sensorData.plot()
```

```python
quatNew = quat
quatNew.plot()
```

```python
from sklearn.preprocessing import MinMaxScaler

np.random.seed(7)
```

```python
train_size = int(len(gtNew[0:20000]) * 0.80)
test_size = len(gtNew[0:20000]) - train_size
```

```python
scalerX = MinMaxScaler(feature_range=(0, 1))
scalerY = MinMaxScaler(feature_range=(0, 1))
```

```python
def create_dataset(datasetInput, datasetOutput, look_back=1):
    dataX, dataY = [], []
    for i in range(len(datasetInput)-look_back-1):
        a = datasetInput[i:(i+look_back), :]
        dataX.append(a)
        dataY.append(datasetOutput[i + look_back, :])
    
    return np.array(dataX), np.array(dataY)
```

```python
sensorData.dropna(how='any', inplace=True)
gt.dropna(how='any', inplace=True)

inputData = sensorData[0:20000]
outputData = gtNew[0:20000]

inputData = scalerX.fit_transform(inputData)
outputData = scalerY.fit_transform(outputData)

trainInput = inputData[0:train_size]
trainOutput = outputData[0:train_size]

testInput = inputData[train_size:len(inputData)]
testOutput = outputData[train_size:len(outputData)]
```

```python
look_back = 1
trainX, trainY = create_dataset(trainInput, trainOutput, look_back)
testX, testY = create_dataset(testInput, testOutput, look_back)

print(trainX.shape)
```

```python
print(trainY.shape)
```

```python
print(testInput.shape)
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], 6, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 6, testX.shape[1]))
```

```python
model = Sequential()
model.add(LSTM(256, input_shape=(6, look_back), return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(256, input_shape=(6, look_back)))
model.add(Dropout(0.3))
model.add(Dense(256))
model.add(Dense(128))
model.add(Dense(3))

#adamOpt = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
#sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='mean_squared_error', optimizer='rmsprop', metrics=['accuracy'])
model.fit(trainX, trainY, epochs=400, batch_size=128, validation_split=0.2, verbose=1, shuffle=True)
```

```python
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
# invert predictions
print(trainPredict.shape)

trainPredict = scalerY.inverse_transform(trainPredict)
trainY = scalerY.inverse_transform(trainY)
testPredict = scalerY.inverse_transform(testPredict)
testY = scalerY.inverse_transform(testY)
```

```python
print(inputData.shape)
print(outputData.shape)
```

```python
import math
from sklearn.metrics import mean_squared_error

trainScore = (mean_squared_error(trainY, trainPredict))
print('Train Score: %.9f MSE' % (trainScore))
testScore = (mean_squared_error(testY, testPredict))
print('Test Score: %.9f MSE' % (testScore))
```

```python
trainPredictPlot = np.empty_like(outputData)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

testPredictPlot = np.empty_like(outputData)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(outputData)-1, :] = testPredict
```
